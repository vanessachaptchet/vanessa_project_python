```python
df2=df.groupby("Continent")["2020 Population"].mean()



plt.figure(figsize=(10,8))
ax = df3.plot(kind = "bar", colormap= "Reds")

for p in ax.patches:
    ax.annotate(
        str(round(p.get_height(), 0)),
        (p.get_x() + p.get_width()/2, p.get_height()),
        ha="center",
        va = "bottom"
        
    )

plt.title("Average Population in 2020")
plt.show()
```

```python
years=["2022 Population","2020 Population","2015 Population","2010 Population","2000 Population"]
df4 = df[years].mean().sort_values(ascending=True)
ax1=df4.plot(kind = "bar", figsize=(10,6), colormap="viridis")

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")
        
plt.title("Average Population by Years")
plt.show()
```

```python
years = ["2022 Population", "2020 Population", "2015 Population", "2010 Population", "2000 Population"]

fig, axes = plt.subplots(3, 2, figsize=(14, 12))
axes = axes.flatten()

for i, year in enumerate(years):
    df.groupby("Continent")[year].sum().plot(
        kind="bar",
        ax=axes[i],
        title=year,
        colormap="viridis"
    )

for j in range(len(years), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
years = ["2022 Population", "2020 Population", "2015 Population", "2010 Population", "2000 Population"]

fig, axes = plt.subplots(3, 2, figsize=(14, 12))
axes = axes.flatten() # transform the grilled to list 

for i, year in enumerate(years):
    ax = df.groupby("Continent")[year].sum().plot(
        kind="bar",
        ax=axes[i],
        title=year,
        colormap="viridis"
    )

    # add values on the plots
    for container in ax.containers:
        ax.bar_label(container, fmt="{:,.0f}" , padding=3)

# remove the 6 plots
for j in range(len(years), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout() # to avoid that the pictures overlap
plt.show()
```

```python
# 1. Filter only Africa
africa_df = df[df["Continent"] == "Africa"]

# 2. Get top 10 most populated countries in Africa (2022)
africa_top10 = africa_df.nlargest(10, "2022 Population")

# 3. Display the result
#print(africa_top10[["Country/Territory", "2022 Population"]])

# 4. Create the bar chart
ax = africa_top10.plot(
    kind="bar",
    x="Country/Territory",
    y="2022 Population",
    figsize=(10,6),
    colormap="viridis"
)

# 5. Add values on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt="{:,.0f}", padding=3)

# 6. Add titles and labels
plt.title("Top 10 Most Populated Countries in Africa (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)

# 7. Show the plot
plt.show()
```

```python
# 1. Filter only Africa
africa_df = df[df["Continent"] == "Africa"]

# 2. Get top 10 most populated countries in Africa (2022)
africa_top10l = africa_df.nlargest(10, "2022 Population")
africa_top10s = africa_df.nsmallest(10, "2022 Population")

fig, axes = plt.subplots(1,2, figsize=(14,6))

# 3. Display the result
#print(africa_top10[["Country/Territory", "2022 Population"]])

# 4. Create the bar chart
ax1 = africa_top10l.plot(
    kind="bar",
    x="Country/Territory",
    y="2022 Population",
    ax=axes[0],
    title=" The most populated Country in 2022",
    #figsize=(10,6),
    colormap="viridis")


# 5. Add values on top of bars
for container in ax1.containers:
    ax1.bar_label(container, fmt="{:,.0f}", padding=3)

# 4. Create the bar chart
ax2 = africa_top10s.plot(
    kind="bar",
    x="Country/Territory",
    y="2022 Population",
    ax=axes[1],
    title="The least populated Country in 2022")
    #figsize=(10,6),
   # colormap="Blues")

for container in ax2.containers:
    ax2.bar_label(container, fmt ="{:,.0f}", padding=3)
# 6. Add titles and labels
#plt.title("Top 10 Most Populated Countries in Africa (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)

# 7. Show the plot
plt.show()
```
